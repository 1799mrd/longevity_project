# Небольшая практика работы с cobrapy escher и cameo


### Информация про библиотеки


Cobrapy - это библиотека Python, которая используется в работе с метаболическими сетями и путями (моделями обмена веществ). Она используется для анализа и моделирования метаболических сетей, а также для прогнозирования и оптимизации метаболических путей.

Escher - это библиотека Python, которая предоставляет инструменты для визуализации и анализа метаболических путей. Она позволяет создавать интерактивные диаграммы путей, отображать метаболиты, реакции и гены, а также проводить анализ и модификацию путей.

Cameo - это библиотека Python, которая используется для анализа и моделирования метаболических путей. Она может быть использована для прогнозирования обменных процессов, оптимизации процессов производства и анализа метаболической функции.

### Работа с человечекими данными с cobrapy
Данные по человеку скачала на http://bigg.ucsd.edu/models/Recon3D
Далее загрузила в блокнот модель в формате .xml

```python
from cobra.io import read_sbml_model
model = read_sbml_model("Recon3D.xml")
```

```python
model
```

```python
print(len(model.metabolites))  # Количество метаболитов
print(len(model.reactions))  # Количество реакций
print(len(model.genes))
```

```python
reactions = model.reactions
for reaction in reactions:
    print(reaction.id, reaction.reaction)
```

```python
metabolites = model.metabolites
for metabolite in metabolites:
    print(metabolite.id, metabolite.name)
```
